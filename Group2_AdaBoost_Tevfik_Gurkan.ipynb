```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Read and view dataset
df = pd.read_csv("Spotify_Dataset/data.csv")
df.shape
```

```python
# Set aside test set generation
test_set = df[df.popularity == 0]
test_set
```

```python
# Separate training and validation sets
train_and_validation = df[df["popularity"] != 0]
train_and_validation.shape
```

```python
# View new distribution
sns.histplot(train_and_validation, x = "popularity")
```

```python
# Separate target variable
y = train_and_validation.popularity
# Keep only numeric attributes for PCA
X_numeric = train_and_validation.drop(['popularity', 'id', 'release_date', 'name', 'artists'], axis=1)
X_numeric.shape
```

```python
# Standardization of the data
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(X_numeric)
X_scaled = ss.transform(X_numeric)
X_scaled
```

```python
from sklearn import decomposition
# Run PCA
pca = decomposition.PCA()
# Store PCs
x_pca = pca.fit(X_scaled).transform(X_scaled)
# Display explained variance
pca.explained_variance_ratio_
```

```python
df = pd.read_csv("spotify_train.csv", index_col=0)
df.shape
```

```python
y = df.popularity
X = df.drop(["popularity"], axis = 1)
X.head()
```

```python
# Standardize the data
ss = StandardScaler()
ss.fit(X)
X_scaled = ss.transform(X)
# View data after transformation
X_scaled
```

```python
from sklearn.model_selection import train_test_split
# Train vs validation split
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.30, random_state=123, stratify = y)
```

```python
X_train.shape
```

**ADABOOST CLASSIFIER:**

**Base Model:**

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import metrics
# Build and fit
ab = AdaBoostClassifier()
ab.fit(X_train, y_train)

# Make predictions
pred = ab.predict(X_train)
pred2 = ab.predict(X_val)

# Calculate accuracy score
print("Training Accuracy: " + str(metrics.accuracy_score(y_true = y_train, y_pred = pred)))
print("Testing Accuracy: " + str(metrics.accuracy_score(y_true = y_val, y_pred = pred2)))

# Balanced Accuracy
print("Training BAccuracy: " + str(metrics.balanced_accuracy_score(y_true = y_train, y_pred = pred)))
print("Testing BAccuracy: " + str(metrics.balanced_accuracy_score(y_true = y_val, y_pred = pred2)))
```

```python
# Create and view confusion matrix
print("Confusion Matrix:")
cm = metrics.confusion_matrix(y_true = y_val, y_pred = pred2)
print(cm)
sns.heatmap(cm/np.sum(cm), annot=True)
```

**Grid Search with Balanced Accuracy:**

```python
from sklearn.model_selection import GridSearchCV
# Perform grid search where you check for combinations of these hyper-parameters.
parameters = {'learning_rate':[0.0001,0.001,0.01,0.1,0.5,1,5,10,100], 
              'n_estimators':range(10,210, 20)
             }
# Classifier object
ab = AdaBoostClassifier()
# Grid Search
gs = GridSearchCV(ab, parameters, verbose=1, cv=5, n_jobs=-1,  scoring = "balanced_accuracy")
# Fit the search
gs.fit(X_train, y_train)
# Print results
print(gs.best_params_)
print(gs.best_score_)
```

**Upsample Non-majority Classes**

```python
print(df.head)
# Display number of cases for each quartile
print(len(df[df.popularity == 'q1']))
print(len(df[df.popularity == 'q2']))
print(len(df[df.popularity == 'q3']))
print(len(df[df.popularity == 'q4']))

# Set majority class
majority = len(df[df.popularity == 'q2'])
```

```python
from sklearn.utils import resample

# Upsample q1
q1_set = df[df.popularity == 'q1']
q1_upsample = resample(q1_set, replace = True, n_samples = majority, random_state = 123)

# Upsample q3
q3_set = df[df.popularity == 'q3']
q3_upsample = resample(q3_set, replace = True, n_samples = majority, random_state = 123)

# Upsample q4
q4_set = df[df.popularity == 'q4']
q4_upsample = resample(q4_set, replace = True, n_samples = majority, random_state = 123)
```

```python
q2_set = df[df.popularity == 'q2']

# Combine upsampled with majority class
upsampled_set = pd.concat([q1_upsample, q2_set, q3_upsample, q4_upsample])

upsampled_set.shape
```

```python
# Separate sets
y2 = upsampled_set.popularity
X2 = upsampled_set.drop(["popularity"], axis = 1)

# Train test split
X_train2, X_val2, y_train2, y_val2 = train_test_split(X2, y2, test_size=0.30, random_state=123, stratify = y2)
```

**Upsampled Base Model:**

```python
# Build and fit
ab = AdaBoostClassifier()
ab.fit(X_train2, y_train2)

# Make predictions
pred = ab.predict(X_train)
pred2 = ab.predict(X_val)

# Calculate accuracy score. What are the training and testing accuracies of the model?
print("Training Accuracy: " + str(metrics.accuracy_score(y_true = y_train, y_pred = pred)))
print("Testing Accuracy: " + str(metrics.accuracy_score(y_true = y_val, y_pred = pred2)))
```

```python
print("Confusion Matrix:")
cm = metrics.confusion_matrix(y_true = y_val, y_pred = pred2)
print(cm)
sns.heatmap(cm/np.sum(cm), annot=True)
```

**Upsampled Model with First Grid Search Results:**

```python
# Build and fit
ab = AdaBoostClassifier(learning_rate=1, n_estimators=170)
ab.fit(X_train2, y_train2)

# Make predictions
pred = ab.predict(X_train)
pred2 = ab.predict(X_val)

# Calculate accuracy score. What are the training and testing accuracies of the model?
print("Training Accuracy: " + str(metrics.accuracy_score(y_true = y_train, y_pred = pred)))
print("Testing Accuracy: " + str(metrics.accuracy_score(y_true = y_val, y_pred = pred2)))
```

```python
print("Confusion Matrix:")
cm = metrics.confusion_matrix(y_true = y_val, y_pred = pred2)
print(cm)
sns.heatmap(cm/np.sum(cm), annot=True)
```

**Undersample Non-minority Classes:**

```python
minority = len(df[df.popularity == 'q4'])

# Undersample q1
q1_set = df[df.popularity == 'q1']
q1_undersample = resample(q1_set, replace = False, n_samples = minority, random_state = 123)

# Undersample q3
q3_set = df[df.popularity == 'q3']
q3_undersample = resample(q3_set, replace = False, n_samples = minority, random_state = 123)

# Undersample q2
q2_set = df[df.popularity == 'q2']
q2_undersample = resample(q2_set, replace = False, n_samples = minority, random_state = 123)
```

```python
q4_set = df[df.popularity == 'q4']

# Combine undersampled with minority class
undersampled_set = pd.concat([q1_undersample, q2_undersample, q3_undersample, q4_set])

undersampled_set.shape
```

```python
# Separate sets
y3 = undersampled_set.popularity
X3 = undersampled_set.drop(["popularity"], axis = 1)

# Train test split
X_train3, X_val3, y_train3, y_val3 = train_test_split(X3, y3, test_size=0.30, random_state=123, stratify = y3)
```
